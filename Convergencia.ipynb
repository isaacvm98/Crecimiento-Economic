```python
data['gdppc']=data['rgdpo']/data['pop']
data[['gdppc_growth','pop_growth']]=data.sort_values(['year']).groupby('country')[['gdppc','pop']].pct_change()
data['ln_gdppc']=np.log(data['gdppc'])
```
