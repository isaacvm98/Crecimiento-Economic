```python
ak = pd.read_excel(r"C:\Users\isaac\Documentos\Eco\C.Econ\pwt91.xlsx", sheet_name = "Data")
```

```python
ak = ak[["country","year","csh_i","rgdpe","pop"]]
```

```python
ak['GDPPPC']=ak['rgdpe']/ak['pop']
ak['GDPPPC_Growth']= ak.sort_values(['year']).groupby('country')[['GDPPPC']].pct_change()
ak['S_Growth']= ak.sort_values(['year']).groupby('country')[['csh_i']].pct_change()
```

```python
ak = ak.dropna()
```

```python
ak= ak[['country','year','S_Growth','GDPPPC_Growth','csh_i']]
```

```python
ak.head()
```

```python
import numpy as np
from scipy.optimize import curve_fit
```

```python
s_mean = ak.groupby(ak['country'])['csh_i'].mean()
y= s_mean
```

```python
gdpppc_mean = ak.groupby(ak['country'])['GDPPPC_Growth'].mean()
x = gdpppc_mean
```

```python
m, b = np.polyfit(x, y, 1)
m,b
```

```python
plt.style.use('default')
plt.figure(figsize =(15,8))
plt.scatter(gdpppc_mean,s_mean)
plt.plot(x, m*x + b,label='RegresiÃ³n',color = 'red')
print ('Los pÃ¡rametros de la regresiÃ³n son pendiente = {} , ordenada al origen = {}'.format(m,b))

plt.legend()
```
